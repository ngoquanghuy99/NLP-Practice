```python
import nltk
import urllib.request
```

```python
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/Nate_Ruess')
```

```python
print("result code: {}".format(str(response.getcode())))
# print 200, that means "that's ok!"
```

```python
text = response.read()
```

```python
print(text)
```

```python
# the above page is about Nate Ruess (my fav artist)
# we'll see whether our code is able to identify the page's content correctly

# we'll use Beautiful Soup - a library just for pulling data out of HTML and XML files
# we'll clean our webpage text of HTML tags

from bs4 import BeautifulSoup
soup = BeautifulSoup(text, 'html5lib')
clean_text = soup.get_text(strip = True)
print(clean_text)
```

```python
# convert the clean text into tokens
tokens = clean_text.split()
# we can use another way:
# tokens = [t for t in clean_text.split()]
```

```python
print(tokens)
```

## Counting word Frequency


A function named FreqDist() will count for us. Also, we will remove stop words (a, at, the, etc) as they will hamper our word freq counting.

```python
# Removing Storpwords in English
from nltk.corpus import stopwords
sr = stopwords.words('english')
clean_tokens = tokens[:]
for t in tokens:
    if t in sr:
        clean_tokens.remove(t)
# Counting freqency
freq = nltk.FreqDist(clean_tokens) ## this freq variable is a Dictionary
```

```python
# see how many times the word "Nate" appeared
print(freq["Nate"])
```

```python
# print pairs of keys and values
for key in freq:
    print(str(key) + ': ' + str(freq["key"]))

# another way: 

### for key, value in freq.items():
###    print(str(key) + ': ' + str(value))
```

```python
# plotting
freq.plot(20, cumulative=False)
```
